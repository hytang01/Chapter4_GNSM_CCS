```python
# Open a file in write mode
fileName = os.getcwd()+"/"+keyword+".DAT"
with open(fileName, 'w') as file:
    file.write(keyword+'\n')
    # Loop through the range of numbers
    for k in range(grids_model_b[2]):
        for j in range(grids_model_b[1]):
            for i in range(grids_model_b[0]):
                # Write the line to the file
                file.write(f'{var_model_b[i][j][k]}\n')
    file.write('/\n')
```

```python
fileName = os.getcwd()+"/"+keyword+".DAT"
size_x = grids_model_b[0]
size_y = grids_model_b[1]
size_z = grids_model_b[2]

var = np.zeros(size_x*size_y*size_z)

with open(fileName) as f:
    lines = f.readlines()
for i in range(len(lines)-2):
    var[i] = math.log(float(lines[i+1]))
    
var = var.reshape((size_x,size_y,size_z),order='F')
```

```python
for i in range(var.shape[2]):
    print('layer No.'+str(i+1))
    plt.figure()
    plt.imshow(var[:,:,i])
    plt.set_cmap('jet')
    plt.colorbar()
    plt.show()
```
